# Week1 (Non-technical) ML Task

## 1. Identify an ML Problem
**Problem:**  Predict whether a student will pass the final exam based on study hours, attendance, and past quiz scores.

## 2. Classify the ML Type
**Type :** Supervised Learning  
**Explanation:**  We have historical records that include input features (study hours, attendance, quiz scores) and a label (passed/failed). The model will learn the mapping from features to the pass/fail label.

## 3. Outline the ML Workflow
1. **Collect data:** Gather past student records from school database or CSV files.  
2. **Clean & preprocess:** Handle missing values, normalize numeric features, encode categorical fields.  
3. **Feature engineering:** Create average quiz score, assignment completion rate.  
4. **Split data:** Train/validation/test = 70/15/15.  
5. **Model selection:** Try Logistic Regression and Random Forest.  
6. **Train & tune:** Use cross-validation and grid search for hyperparameters.  
7. **Evaluate:** Use accuracy, precision, recall, and confusion matrix.  
8. **Deploy & monitor:** Deploy as a simple API and monitor daily accuracy and data drift.

## 4. Identify One Possible Failure Point
**Failure Point:** Bad / insufficient data  
**How to fix it:**  
Collect more historical records, improve data collection process, or choose a simpler model and use cross-validation to reduce overfitting.
